```
from IPython.core.magic import Magics, cell_magic, magics_class

@magics_class
class NodePlugin(Magics):
    @staticmethod
    def write_file(contents):
        with open('source.js', "w") as f:
            f.write(contents)

    @staticmethod
    def run():
        !node ./source.js

    @cell_magic
    def node(self, line, cell):
        self.write_file(cell)
        self.run()

get_ipython().register_magics(NodePlugin)
```
