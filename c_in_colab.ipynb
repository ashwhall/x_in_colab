```
from os.path import isfile
from IPython.core.magic import Magics, cell_magic, magics_class

@magics_class
class GCCPlugin(Magics):
    @staticmethod
    def write_file(contents):
        with open('source.c', "w") as f:
            f.write(contents)

    @staticmethod
    def compile():
        !rm -f app.out
        !gcc source.c -Wall -o app.out
    
    @staticmethod
    def run():
        if isfile('app.out'):
            !./app.out

    @cell_magic
    def gcc(self, line, cell):
        self.write_file(cell)
        self.compile()
        self.run()

get_ipython().register_magics(GCCPlugin)
```
